```python
x_train,x_test,y_train,y_test=train_test_split(data_full2,y,train_size=0.9,random_state=0)
```

dropping unnecessary data columns


```python
dt=DecisionTreeClassifier()
dt.fit(x_train,y_train)
y_pred_train=dt.predict(x_train)
y_pred_test=dt.predict(x_test)
print('training= ',classification_report(y_train,y_pred_train))
print(classification_report(y_test,y_pred_test))
accuracy_score(y_test,y_pred_test)
```

```python
rf=RandomForestClassifier().fit(x_train,y_train)
y_pred_train2=rf.predict(x_train)
y_pred_test2=rf.predict(x_test)
print(accuracy_score(y_test,y_pred_test2))
print(accuracy_score(y_train,y_pred_train2))
```

```python
abc = AdaBoostClassifier(n_estimators=500,
                         learning_rate=0.1)
abc.fit(x_train,y_train)
y_train_pred3=abc.predict(x_train)
y_test_pred3=abc.predict(x_test)
print(accuracy_score(y_train_pred3,y_train))
accuracy_score(y_test_pred3,y_test)
```

```python
xbg=XGBClassifier().fit(x_train,y_train)
y_train_pred4=xbg.predict(x_train)
y_test_pred4=xbg.predict(x_test)
print(accuracy_score(y_train_pred4,y_train))
accuracy_score(y_test_pred4,y_test)
```

```python
classifier=SVC()
param_grid={'kernel':['rbf','poly','sigmoid'],
           'C':[0.1, 1, 10, 50],
           'gamma':[1, 0.1, 0.01]}
svc_gcv = GridSearchCV(estimator=classifier, param_grid=param_grid, cv=10)

svc_gcv.fit(x_train,y_train)
y_train_pred5=svc_gcv.predict(x_train)
y_test_pred5=svc_gcv.predict(x_test)
print('training accuracy',accuracy_score(y_train,y_train_pred5))
print('testing accuracy',accuracy_score(y_test,y_test_pred5))
```

```python
gaussian_nb = GaussianNB()
gaussian_nb.fit(x_train, y_train)
y_train_pred6=gaussian_nb.predict(x_train)
y_test_pred6=gaussian_nb.predict(x_test)
print('training accuracy',accuracy_score(y_train,y_train_pred6))
print('testing accuracy',accuracy_score(y_test,y_test_pred6))
```

```python
# knn
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(x_train,y_train)
y_test_pred7 = knn.predict(x_test)
y_train_pred7 = knn.predict(x_train)
print('training accuracy',accuracy_score(y_train,y_train_pred7))
print('testing accuracy',accuracy_score(y_test,y_test_pred7))
```
